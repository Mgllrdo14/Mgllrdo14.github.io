```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import ipywidgets
import bqplot
%matplotlib inline
```

```python
Vegas = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00397/LasVegasTripAdvisorReviews-Dataset.csv", delimiter=';')

Vegas
```

```python
@ipywidgets.interact(Hotel=Vegas['Hotel name'].unique())
def hotel_plot(Hotel):
    hotel_subset = Vegas.loc[Vegas['Hotel name']==Hotel]
    plt.figure(figsize=(20, 10))
    #fig, ax = plt.subplots(figsize = (20,15))
    hotel_subset.groupby(['Traveler type'])['Score'].agg(['mean']).plot.bar()
    plt.show()
    
```

Are you planning to go on vacation in the future? Las Vegas, Nevada is a city filled with plenty of unique attractions. Though many activities in Vegas are geared toward an adult audience, there are many activities that are enjoyable for people of all ages. Hotels on the Las Vegas Strip are very large in size, and they have the space to for their own attractions such as spas, arcades, tennis courts, rollercoasters, etc. For example, people traveling with family may want a pool at their hotel and data shows that hotels with a pool perform better (See Appendix A). Some hotel attractions such as a spa don't make much difference when it comes to score ratings (See Appendix B). Going to Vegas is all about the experience so plannig a trip requires preparation and research on hotels and the attractions that are accessible. 

Each hotel is unique in what they have to offer for their guests. That being said, some hotels are more geared to families. Before booking a trip, it is obvioulsy important to think about what events you want to have on your itinerary, but who you are traveling with should be a factor as well. The above visualization displays the average score given to different hotels by different type of traveling parties. The plot allows you to choose to see data from 21 different hotels on the Las Vegas Strip. The different traveling party types represented are business, couples, families, friends, and solo. 

Traveling for a business trip may offer a very different experience than traveling friends, family, or significant others. Based on the visualization, the Wynn in Las Vegas is an enjoyable hotel for family and friends to stay at. On the other hand, the data shows that people traveling for business have a less enjoyable time at the Wynn. Different hotels offer better experiences depending on type of travel. Make sure to consider who you are traveling with when booking a hotel.
